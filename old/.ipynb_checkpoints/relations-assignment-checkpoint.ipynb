```acl2
(defsnapshot todo-1)

(defdata nat-pair
    (list nat nat))

(defdata relation-over-nat
    (listof nat-pair))

(definec is-member (x :nat-pair S :relation-over-nat) :boolean
   
    (if (equal S nil)
        
        (if (equal x nil) t nil)
        
        (if (equal x (first S)) t
            (is-member x (rest S)))))

(check-expect (is-member '(1 2) '((1 2))) t)
(check-expect (is-member nil nil) t)
(check-expect (is-member '(a b) '((a b))) nil)
(check-expect (is-member '(nil nil) '()) t)
(check-expect (is-member '(0 1) '((0 1) (0 2) (0 3) (0))) nil)






















```

# TODO-3 (10 pts)

Now let's think of the relation `R` as a relation from a set (of natural numbers) `A` to itself. For example, the set `A` may be `(1 2 3)` and the relation could be `((1 1) (1 2) (2 3))`.

Write a function `(is-reflexive A R)` that returns true if the relation `R` (which is from `A` to `A`) is reflexive.

```acl2
(defsnapshot todo-3)

(defdata list-of-nat
    (listof nat))


(definec is-reflexive (A :list-of-nat R :relation-over-nat) :boolean
    (if (equal A nil) 
        t
        (if (equal R nil)
            nil
            (if (is-related (first A) (first A) R) 
                (is-reflexive (rest A) R)
                nil))))



            
    
(check-expect (is-reflexive '(1 2 3 4 5) nil ) nil)
(check-expect (is-reflexive '(1 2) '((1 1) (2 2))) t)
(check-expect (is-reflexive '(1) '((1 1) (2 2))) t)
(check-expect (is-reflexive '(1 2 3) '((1 1) (2 2))) nil)
(check-expect (is-reflexive '(1 2 3) '((1 1) (1 2) (2 3))) nil)
(check-expect (is-reflexive '(a b c) '((a a) (b b) (c c))) nil)














```

# TODO-4 (10 pts)

Now, let's write the function `(reflexive-closure A R)` that returns the reflexive closure of `R`, which is a relation from `A` ro `A`.

> Hint: think carefully about the type of your expressions, both the arguments and the return type of the function. Definitely write some unit tests **before** trying to write the code. In fact, use the unit tests to suggest the defining equations. This is a harder function than the ones you've done before, but if you stick to the process, you should be able to do it (and level up as a programmer).

```acl2
(defsnapshot todo-4)








(definec reflexive-closure (A :list-of-nat R :relation-over-nat) :relation-over-nat
    (if (equal A nil)
        R
        (if (is-reflexive A R)
            R
            (if (is-related (first A) (first A) R)
                (reflexive-closure (rest A) R)
                (cons (list (first A) (first A)) (reflexive-closure (rest A) R))))))
 









```

## TODO-6 (10 points)

Stop and take a moment to think about the function `(is-reflexive A R)`. Why does this function need both `A` and `R`? Notice that they serve different purposes. The function essentially "loops over" the elements of `A`, and then for each element it performs some test in `R`.

What if we want to loop over all the elements of `R`, and for each such element we need to perform a test on all of `R`? If you ponder that question for a bit, you will understand why this next function has two arguments. Note that we define a general `(is-symmetric-aux R1 R)`, but that it's `(is-symmetric-aux R R)` that determines whether `R` is symmetric!

Write the function `(is-symmetric-aux R1 R)`, and then use it to define `(is-symmetric R)`.

```acl2
(defsnapshot todo-6)



(definec is-symmetric-aux (R1 :relation-over-nat R :relation-over-nat) :boolean
        (if (equal R1 nil)
            t
            (if (equal R nil)
            nil
                (if (is-member (first R1) R)
                    (if (is-related (second (first R1)) (first (first R1)) R)
                    (is-symmetric-aux (rest R1) R)
                        nil)
                    nil))))

(definec is-symmetric (R :relation-over-nat) :boolean
        (is-symmetric-aux R R))
                    















```

## TODO-7 (10 points)

Now write the function `(symmetric-closure R)` that finds the symmetric closure of the relation `R`. Try not to add any unnecessary pairs in the closure, so that the result is as small as possible.

```acl2
(defsnapshot todo-7)




(definec symmetric-closure (R :relation-over-nat) :relation-over-nat
    (if (equal R nil)
        R
        (if (is-symmetric R)
            R
            (if (is-symmetric (rest R))
                (append R '((second (first R)) (first (first R))))
            
                (append (symmetric-closure (rest R)) 
                        '((second (first R)) (first (first R))))
))))
            
            
            



(check-expect (symmetric-closure '((0 1) (1 2) (3 4))) '((0 1) (1 0) (2 1) (1 2) (4 3) (3 4)))









```

## TODO-10 (10 points)

We are now going to tackle the transitive closure of a relation `R`. Remember that this can be found as 
$$R^+ = R \cup R\circ R \cup R \circ R \circ R \dots$$
where $A\circ B$ is relation composition.

The definition of $R^+$ above has a major flaw, because it is an infinite formula. In general, we cannot do this, but in this case we can easily get around it, because if the cardinality of $R$ is $n$, then we can stop after we take $n$ compositions. (Think about why this is true.) So we really have just
$$R^+ = R \cup R\circ R \cup R \circ R \circ R \dots \cup \underbrace{R \circ R \circ R \dots \circ R}_{n=|R|\text{ times}}$$
