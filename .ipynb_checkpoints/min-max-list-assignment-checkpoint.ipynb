## TODO-1 (10 points)

The first step is to define a function `(sumlist L)` that returns the sum of all the elements in the list `L`. E.g.,

    (sumlist '(1 2 3 4)) = 1+2+3+4

> Note: You may assume that the list `L` is a list of natural numbers. (It's OK if you do something else, like a list of rational numbers, but my solution treats `L` as a list of naturals.

> Note: Don't forget to write unit tests for all functions. The unit tests should help you write the code, particularly in writing the defining equations.


```acl2
(defsnapshot from-the-top)
(defsnapshot todo-1)


(definec sumlist (ls :nat-list) :nat
    (if (= 0 (len ls))
        0
        (+ (first ls)(sumlist (rest ls)))
    )
)

(check-expect (sumlist '(1 2 3)) 6)
(check-expect (sumlist '(1 1 1 1 1 1)) 6)
(check-expect (sumlist '(0 1 0 1)) 2)
(check-expect (sumlist '(99 0 1 000 5)) 105)
(check-expect (sumlist '(0 0 1)) 1)
(check-expect (sumlist '(0 0 0)) 0)

 ;; true even by prop. reasoning
(test? (implies (nat-listp '(a b c))
                (= (sumlist '(a b c))(+ (third '(a b c))(second '(a b c))(first '(a b c))))
        )
)






```

## TODO-2 (10 points)

You know that `:nat` represents the type of natural numbers. Define a new data type `:optnat` that is either a natural number or the special constant `NIL`. Think of `NIL` as `NaN` in computer arithmetic (and `NaN` stands for "not a number"). Be sure to write unit tests for your data type, so that you know it really does represent what you think it does.

```acl2
(defsnapshot todo-2)

;; described opt-nat data type
(defdata opt-nat (oneof NIL nat))

;; list of opt-nat
(defdata opt-nat-list (listof opt-nat))

;; working with list of opt-nat's
(definec opt-sumlist (ls :opt-nat-list) :nat
    (if (= 0 (len ls))
        0
        (if (not (natp (first ls)))
            (+ 0 (opt-sumlist (rest ls)))
            (+ (first ls)(opt-sumlist (rest ls)))
        )
    )
)

;; properties of opt-nat
(check-expect (opt-natp nil) T)
(check-expect (opt-natp 'X) NIL)
(check-expect (opt-natp 1) T)

;; properties of opt-nat when in lists
(check-expect (opt-nat-listp '(2 1 0 nil)) T)
(check-expect (opt-nat-listp '(beckham 2 1 0 nil)) NIL)
(check-expect (opt-nat-listp '()) T)

;; properties of operating on opt-nat
(check-expect (opt-sumlist '(1 nil 3)) 4)
(check-expect (opt-sumlist '(nil nil 0 nil)) 0)
(check-expect (opt-sumlist '(0 1 0 1)) 2)

;; nat is a subset of opt-nat
(test? (implies (natp a)(opt-natp a)))

;; opt-nat is a superset of nat
(test? (implies (= a NIL)(and (opt-natp a)(not(natp a)))))

;; nat subset opt-nat continued
(test? (implies (nat-listp a)(opt-nat-listp a)))




```

## TODO-3 (10 points)

Now write the function `(minlist L)` that returns the smallest number in the list `L`. E.g.,

    (minlist (list 2 4 6 8)) = 4

If the list `L` does not have any elements, then `minlist` should return `NIL` (to indicate an error).


```acl2
(defsnapshot todo-3)

(definec minlist (ls :nat-list) :int ;acl2 did not like using opt-nat here, so *we can't return NIL* However,
    (if (= (len ls) 1)               ;nat-list can't contain -1, so if -1 is returned then it was an error
        (first ls)
        (if (= (len ls) 0)
            -1 
            (min (first ls) (minlist (rest ls)))
        )
    )
)

(check-expect (minlist '(99 50 10)) 10)
(check-expect (minlist '(10 50 99)) 10)
(check-expect (minlist '()) -1)
(check-expect (minlist '(1 1)) 1)
(check-expect (minlist '(11 7 5 3 0)) 0)

(test? (implies (nat-listp '(a b))(= (minlist '(a b))(min a b))))





```

## TODO-4 (10 pts)

Now implement `(maxlist L)`, which should be almost identical to `minlist`.


```acl2
(defsnapshot todo-4)

(definec maxlist (ls :nat-list) :int ;acl2 did not like using opt-nat here, so *we can't return NIL* However,
    (if (= (len ls) 1)               ;nat-list can't contain -1, so if -1 is returned then it was an error
        (first ls)
        (if (= (len ls) 0)
            -1 
            (max (first ls) (maxlist (rest ls)))
        )
    )
)

(check-expect (maxlist '(99 50 10)) 99)
(check-expect (maxlist '(10 50 99)) 99)
(check-expect (maxlist '()) -1)
(check-expect (maxlist '(1 1)) 1)
(check-expect (maxlist '(11 7 5 3 0)) 11)

(test? (implies (nat-listp '(a b))(= (maxlist '(a b))(max a b))))






```

## Aside

You may be feeling a little strange about TODO-6. I said "`L` is a list of naturals", not "`L` is a non-empty list of naturals". So what happens when `L` is empty? In that case, both `minlist` and `maxlist` return `NIL`, so is this true or false:

    (<= NIL NIL)

Your first instinct might be that it's undefined, since `<=` compares numbers and `NIL` is not number. But ACL2 functions **always** return a value, no matter what. So it's not undefined. It's *something*. What ACL2 in most of these cases is to treat each invalid input as some sort of default value. So when it sees `NIL`, it treats that input as if it was `0`. Since `(<= 0 0)` is true, that's also the value of `(<= NIL NIL)`.

So, what do you think is the value of

    (<= NIL 'b)
    (<= 'a 'b)
    (<= 'b 'a)

```acl2
(defsnapshot ec-3)

(definec avglist (ls :nat-list) :rational 
    (if (= 0 (len ls))
        -1
        (/ (sumlist ls) (len ls))
    )
)

(test? (implies (nat-listp ls)(<= (minlist ls)(avglist ls)))) ; ACL2 couldn't prove this with thm

(definec min-less-all (ls :nat-list min :int) :boolean
    (if (= 0 (len ls))
        T
        (if (< (first ls) min)
            NIL
            (min-less-all (rest ls) min)
        )
    )
)

(definec min-less-all-aux (ls :nat-list) :boolean
    (if (= 0 (len ls))
        T
        (min-less-all ls (minlist ls))
    )
)

;; (preliminary) 'proof succeeded'
(thm (implies (nat-listp ls) (min-less-all-aux ls)))


;; this is as far as I could get, I don't understand ACL2's proof methods enough
;; to know what hints it needs so I just guessed that proving the above thm
;; might give the assistance needed

(thm (implies (and (min-less-all-aux ls)(nat-listp ls))
                (<= (* (len ls)(minlist ls))(sumlist ls))
        )
)






```
